```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

# Training dataset

```python
work_dir = os.getcwd()
data_dir ="../../for_students/data_v2"
train_dir = os.path.join(data_dir, 'training')
train_str = os.path.join(train_dir, 'training_{}.csv')
train_df_1 = pd.read_csv(train_str.format('1'))
train_df_2  = pd.read_csv(train_str.format('2'))
train_df_3  = pd.read_csv(train_str.format('3'))
train_df_4  = pd.read_csv(train_str.format('4'))
train_df_5  = pd.read_csv(train_str.format('5'))
train_df_100  = pd.read_csv(train_str.format('100'))
# drop 'request_count' column in dataset_100
train_df_100 = train_df_100.drop("request_count", axis=1)
train_dfs = [train_df_1,train_df_2,train_df_3,train_df_4,train_df_5,train_df_100]
```

```python
# train_df_100
```

```python
for df in train_dfs:
    plt.figure()
    df['kpi_value'].plot()
    df["anomaly_label"][df["anomaly_label"]==1].plot(style="o")
```
