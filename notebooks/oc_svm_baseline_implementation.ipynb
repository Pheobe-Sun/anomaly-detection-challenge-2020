### Some Simple Baselines (Novelty Detection OC - SVM)

https://scikit-learn.org/stable/modules/generated/sklearn.svm.OneClassSVM.html


https://dl.acm.org/doi/10.1162/089976601750264965



```python
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.metrics import f1_score, accuracy_score
```

```python
def OC_SVM_Baseline(train_df): # input a dataframe object as per data loading
    
    y_true = train_df[['label']].values #true label
    
    clf = OneClassSVM(gamma='auto', kernel='rbf').fit(train_df[['value']].values) # OC-SVM
    
    y_pred = clf.predict(train_df[['value']].values) #Predicted Value from OC-SVM
    
    y_pred = np.array([0 if y==1 else 1 for y in list(y_pred)]) #to be in-line with sklearn output
    
    return(print('F1 score of train_df is {} with {} anomalies detected, true amount of anomalies is {}'.format(f1_score(y_true,y_pred), list(y_pred).count(1), list(y_true).count(1)))) 
        
        #print(f1_score(y_true,y_pred)), print(list(y_pred).count(1))) # return F1 Measure and number of anomoly points. 
```

```python
for dataframe in [train_df_1, train_df_2, train_df_3 ]:
    OC_SVM_Baseline(dataframe)
```
